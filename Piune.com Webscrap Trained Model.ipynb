```python
!pip install pandas
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://pixune.com/blog/how-much-does-a-3d-animation-cost/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

# Extract all paragraphs
all_paragraphs = soup.find_all('p')
data = [{'text': para.get_text(strip=True)} for para in all_paragraphs]

# Create a pandas DataFrame
df = pd.DataFrame(data)
print(df)


# Extracting pricing info (example - may need adjustment based on website structure)
pricing_info = []
for para in all_paragraphs:
  text = para.get_text(strip=True)
  if '$' in text:  # Assuming pricing is indicated by '$' sign
    pricing_info.append({'pricing': text})

pricing_df = pd.DataFrame(pricing_info)
print("\nPricing Information:")
print(pricing_df)


# Save the DataFrames to CSV files
df.to_csv('extracted_all_data.csv', index=False)
pricing_df.to_csv('extracted_pricing_data.csv', index=False)
```
